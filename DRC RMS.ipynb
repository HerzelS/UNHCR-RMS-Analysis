# INTRODUCTION

This data pipeline is looking at data for the RBSA Livelihoods evaluation. However, the functions developed in this data pipeline can be reused in other evaluations and analytical work. A private github repository has been developed to house all the files used and can be made accessible to UNHCR Evaluation Office colleagues. The main focus of this data pipeline  is to:

* **Ensure the utilisation of reproducible data, making it easy to maintain institutional memory, share history with relevant colleagues and evaluation teams, document the thought process, update and correct errors**
* **Provide on one go; a pipeline that imports, cleans, wrangles, analyses and plots graphs all on one click within the R environment - no need to go to MS Excel or MS Word, everything is done on this document**
* **Reproducibility - this pipeline ensures that data is reproducible and can be run by other colleagues, just fork the Git Repository and run this document, no need to re-do the data preparation multiple times.**

More specific to the RBSA Livelihoods Evaluation, this data pipeline cleans and wrangles data from the following:

* 7 RMS Data - the data covers the period 2022-2023 and was extracted from RIDL.

The data pipeline does the following:

1. Extract, clean and wrangle the three datasets mentioned above.
2. Join the three datasets into one complete data frame.
4. Export different cleaned datasets (optional)


### Indicators under exploration

The main indicators for extranction are:

1. Indicator 13.1 Proportion of people with an account at a bank or other financial institution or with a mobile-money-service provider

2. Indicator 13.2 Proportion of people who self-report positive changes in their income compared to previous year

3. Indicator 13.3 Proportion of people (working age) who are unemployed

However, the extraction will also include several demographic data and other related outcome data that can be used to analyse the interaction between livelihood outcomes and other outcome areas tracked by the RMS surveys.
